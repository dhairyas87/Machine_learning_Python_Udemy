```python
# Simple Linear Regression for a machine learning model Python.

#Libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#reading the data

dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values


#Splitting into test and train

from sklearn.model_selection import train_test_split
X_Train,X_Test,y_Train,y_Test =  train_test_split(X,y,test_size = 0.3, random_state=0)

print(X_Train)
print(X_Test)
print(y_Train)
print(y_Test)
```

```python
#Simple linear regression

#importing Linear Regression class from linear mode

from sklearn.linear_model import LinearRegression

#Importing the linear regression and training the training set with linear regression using fit method.
regressor =  LinearRegression()
regressor.fit(X_Train,y_Train)

#Now predicting the linear regression method.
y_pred = regressor.predict(X_Test)

print(y_pred)


#Visualizing the Training set data 


plt.scatter(X_Train,y_Train,color='red')
plt.plot(X_Train,regressor.predict(X_Train),color='blue')
plt.title('Salary vs Experience --Training set')
plt.xlabel('Years of Experience')
plt.ylabel('salary')
plt.show()

```

```python
plt.scatter(X_Test,y_Test,color='red')
plt.plot(X_Test,y_pred,color='blue')
plt.title('Salary vs Experience --Training set')
plt.xlabel('Years of Experience')
plt.ylabel('salary')
plt.show()
```
